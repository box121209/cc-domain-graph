# Playing with unicode

```python
alphabet = range(1160, 1260)

for a in alphabet:
    b = unichr(a)
    c = ord(b)
    print(b + "\t%4d\t%4d\t%s" % (a, c, b.encode("utf-8").encode("hex")))
```

Note that the hex encodings are <b>not</b> the same as the indices in general!

Some examples:

```python
index = [range(35,41),
    range(65,71),
    range(188, 194),
    range(945, 951),
    range(1075, 1081),
    range(1900, 1906),
    range(2335, 2341),
    range(13351, 13357)]

print('---------------------------------------------')
for idx in index:
    print('\t'.join([unichr(x) for x in idx]))
    print('\t'.join([unichr(x).encode("utf-8").encode("hex") for x in idx]))
    print('---------------------------------------------')
```

Dump a string of these characters to a file for testing:

```python
idx = index

str = ''.join([unichr(x[i]) for x in idx for i in range(6)])

f = open('test.txt', 'w')
f.write(str.encode('utf-8'))
f.close()

print(str.encode('utf-8'))
```

Note that the file can be read with <tt>cat</tt>, or viewed in hex with 

<tt>$ cat test.dat | hexdump</tt>

To read back into Python:

```python
with open('test.txt', 'r') as f:
    content = f.read()

print(content)
print(content.encode('hex'))

#print(binascii.hexlify(content))
#print(binascii.hexlify(str.encode('utf-8')))
#print(binascii.unhexlify(binascii.hexlify(content)))
#print(binascii.unhexlify(binascii.hexlify(content)).decode('utf-8'))
#print(binascii.unhexlify(binascii.hexlify(str.encode('utf-8'))).decode('utf-8'))
```

We could use the package <tt>binascii</tt> but we don't need to.

We can also iterate through the byte sequence:

```python
' '.join([b.encode("hex") for b in content])
```

```python
for b in content:
    a = int(b.encode('hex'), 16)  # 16 is the base
    print("%s %4d" % (unichr(a), a))
```

Can we demarcate some of the separate non-Latin scripts?

```python
def hx(a):
    return [x.encode('hex') for x in list(a.encode("utf-8"))]

rg = ['3040','309f'] # hiragana
rg = ['0e00','0e7f'] # thai

for i in range(int(rg[0], 16), int(rg[1], 16)):
    a = unichr(i)
    h = hx(a)
    print(hex(i))
    print(a) 
    print(h)
```

```python
import csv

lookup_file = "./unicode_ranges.txt"
lookup_script = ['' for i in range(1000000)]
lookup_range = {}

with open(lookup_file) as tsv:
    for line in csv.reader(tsv, delimiter='\t'): 
        mn = int(line[0], 16)
        mx = int(line[1], 16)
        lookup_range[line[2]] = [mn, mx+1]
        for i in range(mn, mx+1):
            lookup_script[i] = line[2]

lookup_range
```

```python
def find_scripts(str):
    d = {}
    for c in str:
        s = lookup_script[ord(c)]
        if not s in d.keys(): d[s] = 0
        d[s] += 1
    return d

ex = 'Dingbats'

mn = lookup_range[ex][0]
mx = lookup_range[ex][1]

str = ''.join([unichr(i) for i in range(mn, mx)])
print(str)

print(find_scripts(str))
```

```python
from pprint import pprint

str = ''.join([unichr(i) for i in range(4000)])

pprint(find_scripts(str))
```
